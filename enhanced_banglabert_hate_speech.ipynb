```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="tajuarAkash/enhanced-banglabert-hate-speech")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("tajuarAkash/enhanced-banglabert-hate-speech")
model = AutoModelForSequenceClassification.from_pretrained("tajuarAkash/enhanced-banglabert-hate-speech")
```

```python
# Function to perform inference
import torch
def predict(text):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

    # Move the inputs to the device (GPU or CPU)
    inputs = {k: v.to(model.device) for k, v in inputs.items()}

    # Run the model to get predictions
    with torch.no_grad():
        outputs = model(**inputs)

    # Get probabilities using softmax and extract the class prediction
    probs = torch.softmax(outputs.logits, dim=-1)
    pred = torch.argmax(probs, dim=-1).item()

    # Interpret the result (assuming '1' = Hate, '0' = Non-Hate)
    return {
        'text': text,
        'prediction': 'Hate' if pred == 1 else 'Non-Hate',
        'confidence': probs.max().item(),
        'hate_prob': probs[0][1].item()  # Probability of 'Hate' class
    }

# Example usage:
text = "তুমি একজন vhalo মানুষ।"
result = predict(text)

# Print result
print(result)
```
